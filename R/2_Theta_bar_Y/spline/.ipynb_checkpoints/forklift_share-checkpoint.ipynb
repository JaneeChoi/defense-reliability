```r
# 전동지게차, 2013.1.1~2021.6.29, 일반/긴급대응 데이터

df.1 <- df %>%
        filter(visit_type %in% c("일반대응","긴급대응")) %>%
        filter(date >= '2013-1-1') %>%
        filter(power=="전동") 
df.2 <- df.1 %>%
        select(date,cust_code,manuf,sn,oper_time,visit_type,year_use)
df.3 <- unique(df.2)
df.3 <- na.omit(df.3) 

write_xlsx(df.3,"df.xlsx")
```

```r
# 제조사별 고장건수 분석
a<-as.data.frame(table(df.3$manuf))
a<-a[c(order(a$Freq,decreasing = T)),]
a

b<-a[c(a$Freq>1000),]
b
manu <- as.character(unlist(b$Var1))
manu  # 고장발생 1000건 이상 7개 회사 선택
```

```r
# 제조사별 sn
# clark : manu[1]
m.sn <- df.3 %>% 
        filter(manuf==manu[1]) %>%
        select(sn)
clark <- as.character(unique(m.sn$sn))

# nizziu : manu[2]
m.sn <- df.3 %>% 
        filter(manuf==manu[2]) %>%
        select(sn)
nizziu <- unique(m.sn$sn)

# toyota : manu[3]
m.sn <- df.3 %>% 
        filter(manuf==manu[3]) %>%
        select(sn)
toyota <- unique(m.sn$sn)

# crown : manu[4]
m.sn <- df.3 %>% 
        filter(manuf==manu[4]) %>%
        select(sn)
crown <- unique(m.sn$sn)

# smitomo : manu[5]
m.sn <- df.3 %>% 
        filter(manuf==manu[5]) %>%
        select(sn)
smitomo <- unique(m.sn$sn)

# haister : manu[6]
m.sn <- df.3 %>% 
        filter(manuf==manu[6]) %>%
        select(sn)
haister <- unique(m.sn$sn)

# doosan : manu[7]
m.sn <- df.3 %>% 
        filter(manuf==manu[7]) %>%
        select(sn)
doosan <- unique(m.sn$sn)
```
